```python
import torch
from ultralytics import YOLO
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score
import os
import cv2
import numpy as np

# Load the YOLOv8 model
model = YOLO('/kaggle/input/test-dataset/test/best.pt')

# Paths to your image and label folders
image_folder = '/kaggle/input/test-dataset/test/images'
label_folder = '/kaggle/input/test-dataset/test/labels'

# Function to read labels from a file
def read_labels(label_path):
    try:
        with open(label_path, 'r') as file:
            labels = np.array([line.strip().split() for line in file.readlines()], dtype=float)
            return labels[:, 0].astype(int)  # Return class IDs
    except FileNotFoundError:
        return np.array([])

# Lists to store true and predicted labels
true_labels = []
pred_labels = []

# Loop through each image
for image_name in os.listdir(image_folder):
    image_path = os.path.join(image_folder, image_name)
    label_path = os.path.join(label_folder, image_name.replace('.jpg', '.txt'))

    # Read the image
    image = cv2.imread(image_path)

    # Predict using YOLOv8 with verbose=False to suppress additional output
    results = model.predict(image, save=False, verbose=False)

    # Extract predicted class labels – handle empty predictions
    if results[0].boxes.data.size(0) > 0:
        predicted_classes = results[0].boxes.cls.cpu().numpy().astype(int)
    else:
        predicted_classes = np.array([-1])  # Placeholder for no predictions

    # Read true labels from file
    true_classes = read_labels(label_path)

    # Ensure lengths match by repeating placeholder if needed
    max_len = max(len(true_classes), len(predicted_classes))
    
    # If true_classes are fewer, pad with -1
    if len(true_classes) < max_len:
        true_classes = np.pad(true_classes, (0, max_len - len(true_classes)), constant_values=-1)

    # If predicted_classes are fewer, pad with -1
    if len(predicted_classes) < max_len:
        predicted_classes = np.pad(predicted_classes, (0, max_len - len(predicted_classes)), constant_values=-1)

    # Add true and predicted classes to lists
    true_labels.extend(true_classes)
    pred_labels.extend(predicted_classes)

# Calculate metrics - handle empty arrays
if true_labels and pred_labels:  # Check if both lists are not empty
    precision = precision_score(true_labels, pred_labels, average='weighted', zero_division=1)
    recall = recall_score(true_labels, pred_labels, average='weighted', zero_division=1)
    f1 = f1_score(true_labels, pred_labels, average='weighted', zero_division=1)
    accuracy = accuracy_score(true_labels, pred_labels)

    # Print only the final metrics
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1 Score: {f1:.4f}")
    print(f"Accuracy: {accuracy:.4f}")
else:
    print("No predictions or true labels found.")
```
